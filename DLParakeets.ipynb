```
from google.colab.patches import cv2_imshow
import numpy as np 
  
# Read First Alexandrine Image 
alex1 = cv2.imread('visualization samples/alexandrine1.JPG') 
  
# Read Second Alexandrine Image 
alex2 = cv2.imread('visualization samples/alexandrine2.JPG') 
  
# Read First Bourke Image 
bourke1 = cv2.imread('visualization samples/bourke251.jpg') 
  
# Read Second Bourke Image 
bourke2 = cv2.imread('visualization samples/bourke252.jpg') 

# Read First Budgie Image 
budgie1 = cv2.imread('visualization samples/budgerigar1.JPG') 
  
# Read Second Budgie Image 
budgie2 = cv2.imread('visualization samples/budgerigar5.JPG') 

# Read First Derbyan Image 
derbyan1 = cv2.imread('visualization samples/derbyan4.JPG') 
  
# Read Second Derbyan Image 
derbyan2 = cv2.imread('visualization samples/derbyan5.JPG') 

# Read First Lineolated Image 
line1 = cv2.imread('visualization samples/lineolated1.jpg') 
  
# Read Second Lineolated Image 
line2 = cv2.imread('visualization samples/lineolated2.jpg') 

# Read First Monk Image 
monk1 = cv2.imread('visualization samples/monk1.JPG') 
  
# Read Second Monk Image 
monk2 = cv2.imread('visualization samples/monk2.PNG') 

# Read First Moustache Image 
mous1 = cv2.imread('visualization samples/moustache4.JPG') 
  
# Read Second Moustache Image 
mous2 = cv2.imread('visualization samples/moustache5.JPG') 

# Read First Plain Image 
plain1 = cv2.imread('visualization samples/plain3.jpg') 
  
# Read Second Plain Image 
plain2 = cv2.imread('visualization samples/plain4.jpg') 

# Read First Plum-headed Image 
plum1 = cv2.imread('visualization samples/plum-headed8.JPG') 
  
# Read Second Plum-headed Image 
plum2 = cv2.imread('visualization samples/plum-headed9.JPG')

# Read First Regent Image 
regent1 = cv2.imread('visualization samples/regent3.JPG') 
  
# Read Second Regent Image 
regent2 = cv2.imread('visualization samples/regent4.JPG')

# Read First Ring-necked Image 
ring1 = cv2.imread('visualization samples/ring-necked10.JPG') 
  
# Read Second Ring-necked Image 
ring2 = cv2.imread('visualization samples/ring-necked11.JPG')

# concatanate image Horizontally 
Alex = np.concatenate((alex1, alex2), axis=1) 
Bourke = np.concatenate((bourke1, bourke2), axis=1)
Budgie = np.concatenate((budgie1, budgie2), axis=1)
Derbyan = np.concatenate((derbyan1, derbyan2), axis=1)
Line = np.concatenate((line1, line2), axis=1)
Monk = np.concatenate((monk1, monk2), axis=1)
Mous = np.concatenate((mous1, mous2), axis=1)
Plain = np.concatenate((plain1, plain2), axis=1)
Plum = np.concatenate((plum1, plum2), axis=1)
Regent = np.concatenate((regent1, regent2), axis=1)
Ring = np.concatenate((ring1, ring2), axis=1)
  
cv2_imshow(Alex)
print('Alexandrians')
cv2_imshow(Bourke) 
print('Bourkes')
cv2_imshow(Budgie) 
print('Budgerigars')
cv2_imshow(Derbyan) 
print('Derbyans')
cv2_imshow(Line) 
print('Lineolated')
cv2_imshow(Monk) 
print('Monks')
cv2_imshow(Mous) 
print('Moustaches')
cv2_imshow(Plain) 
print('Plain')
cv2_imshow(Plum) 
print('Plum-headed')
cv2_imshow(Regent)
print('Regents') 
cv2_imshow(Ring)
print('Ring-necked')   
  
cv2.waitKey(0) 
cv2.destroyAllWindows() 
```
